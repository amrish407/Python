### **Logarithm**

The exponent or power to which a base must be raised to yield a given number. Expressed mathematically, x is the logaritm of n to the base b if bx = n, in which case one writes x = logb n.

For example, 23 = 8; therefore, 3 is the logaritm of 8 to base 2, or 3 = log2 8


#### Python functions for computing log:

- math.log(x) -> returns log(x) with base = e
- math.log(x, a) -> returns log(x) with base = a
- math.log2(x)  -> returns log(x) with base = 2

### **Prime** **Number** - sqrt(N) **method**

```python
def isPrime(n):
  """
    This function returns T or F based on primality check of n.
    Time Complexity - O(sqrt(n))
  """

  if n == 1:
    return False

# Search for factors in the range [2, sqrt[n]]
  # sq = n **;(1/2)
  sq = int(math.sqrt(n))

  for i in range(2, sq + 1):
    if n%i == 0:
      return False

  return True
```

```python
isPrime(13)
```

### **AP** - **Arithmetic** **Progrssion**

An AP, also called an arithmetic sequence, is a sequence of numbers which differ from each other by a common difference.

```python
def printAP(a, d, n):

    curr_term = a

    for i in range(curr_term, n+1):
        print(curr_term, end=' ')
        curr_term = curr_term + d
```

```python
# Find nth term
def lastTerm(a,d,n):
  t_n = a + (n-1)*d
  return t_n

# Sum of n terms
def sumOfNTerms(a, d, n):
  sum = (n/2)*(2*a +(n-1)*d)
  return sum
```

### Combinations

math module method -
math.comb(n, r)

```python
math.comb(10, 4)
```

```python
def combination(n, r):
  num = math.factorial(n)
  den = math.factorial(r) * math.factorial(n-r)
  return num/den
```

```python
combination(10, 4)
```

```python
# own function
def decimalToBinary(n):
  l = []
  while n > 0:
    rem = n%2
    l.append(rem)
    n = n//2

# Reverse the List
  l = list(reversed(l))
  return l
```

### **Binary** **to** **Decimal**

Using python inbuilt functionality:

        return int(x, 2)

```python
def binaryToDecimal(x):
    return int(x, 2)
```

```python
binaryToDecimal("100")
```

```python
def binaryToDec(s):
  i = 0
  result = 0

  for digit in s[::-1]:
    result += int(digit) * (2 ** i)
    i = i+1

  return result
```

```python
binaryToDec("100")
```

### **Problem** - Trailing Zeros in Factorial

Given an integer A, return the number of trailing zeros in A!

Note: Your solution should run in logarithmic TC

INPUT: 5

OUTPUT: 1

```python
def trailingZeros(n):
  i = 1
  x = n // 5
  ans = 0
  while x >= 1:
    x = n // (5 ** i)
    ans += x
    i += 1

  return ans

# TC is O(log n) with base 5
```

```python
trailingZeros(30)
```

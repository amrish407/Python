### **Challenge** - Check Palindrome!

Given a string A consisting of lowercase characters.

Check if characters of the given string can be rearranged to form a palimdrome.

Return 1 if it is possible to rearrange the characters of the string A such that it becomes a palindrome else return 0.

Input : "abcde"

Output : 0


Input : "abba"

Output : 1

```python
a = "abba"
```

```python
def solve(a):
  # declare a frequency map (dictionary)
  freq = {}

  for char in a:
    if char in freq:
      freq[char] += 1
    else:
      freq[char] = 1

  #Freq map created

  odd_freq = 0

  for frequency in freq.values():
    if frequency %2 == 1:
      odd_freq += 1

  if odd_freq <= 1:
    return 1
  else:
    return 0

# time complexity O(n)
```

```python
solve("asdf")
```

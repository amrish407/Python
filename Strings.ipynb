## **Function** - **kwargs**

keyword arguments

```python
# kwargs manages arguments which user can enter at runtime
def person(name, age, profession, **kwargs):
  print(name, age, profession)
  print(kwargs)
  print(type(kwargs))
```

```python
person("Peter Parker", 20, "Photographer")
```

```python
#use of kwargs
person("Peter Parker", 20, "Photographer", city="Mumbai", city1="Thane")
```

```python
def hogwarts(house1, house2, house3, house4, *args, **kwargs):
  print("Houses are ", house1, house2, house3, house4)
  print()
  print("Args :", args)
  print()
  print("kwargs :", kwargs)
```

```python
hogwarts(
    "Gryffindor",
    "Ravenclaw",
    "Hufflepuff",
    "Slytherin",
    "Severus Snape",
    "Horace Slughorn",
    "Remus Lupin",
    headmaster="Albus Dumbledore",
    deputy_headmistress="Minerva McGongall"
)
```

### Pattern problem

    A
    BB
    CCC
    DDDD


```python
n = 4
for i in range(1, n+1):
  c = chr(i + ord('A')-1)
  s = (c + " ")*i
  print(s)
```

### **Format** **String**

```python
apple_count = 4
orange_count = 10
guava_count = 5

#generate a string that says, "I have 4 apples, 10 oranges and 5 guavas"
```

```python
s = "I have " + str(apple_count) + " apples, " + str(orange_count) + " oranges and " + str(guava_count)+ " guavas"
s
```

```python
# use this method
s = f"I have {apple_count} apples, {orange_count} oranges and {guava_count} guavas"
s
```

```python
s = "I have {} apples, {} oranges and {} guavas".format(apple_count, orange_count, guava_count)
s
```

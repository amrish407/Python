### **self** **argument**

* The first argument of a method of a class is generally referred as 'self'.

* *self* refers to the calling object.

* This argument is passed implicitly by the Python interpreter and need not be mentioned while calling.

* Similar to *this* keyword in c++, java, ...


````python
class Car:
  def describe(self):
    print("Car name =", self.name)
    print("Car color =", self.color)
    print("Car age =", self.age)

  def isBanned(self):
    #return self.age > 15
    if self.age > 15:
      return True
    else:
      return False
````

````python
benz = Car()
benz.name = "Mercedes"
benz.color = "Silver"
benz.age = 3
````

````python
maruti = Car()
maruti.name = "Maruti 800"
maruti.color = "Red"
maruti.age = 20
````

````python
benz.describe() # while calling the func, we have not passed the arguments
````

````python
maruti.describe()
````

### \_\_init__() **method**

```
-> __init__() is a special method inside a class that allows us to define the
properties/attributes while creating it.

-> It is a special method that is called/invoked automatically when the object
is created. The user need not call it.

-> It is similar to 'Constructors' in C++ and Java

Syntax:

      class Car:
        def __init__(self):
          ...initialization...
```



````python
class Car:
  def __init__(self, name, color, age):
    self.name = name
    self.color = color
    self.age = age

  def describe(self):
    print("Car name =", self.name)
    print("Car color =", self.color)
    print("Car age =", self.age)

  def isBanned(self):
    if self.age > 15:
      return True
    else:
      return False
````

````python
benz = Car("Mercedes", "Silver", 3)

# as with init we have constructor somwe can assign directly
````

````python
benz.name
````

````python
benz.isBanned()
````

Another Example

````python
class Wizard:
  def __init__(self, name, age, spell):
    self.name = name
    self.age = age
    self.spell = spell

  def magic(self):
    print("Cast spell - ", self.spell)
````

````python
harry = Wizard("Harry Potter", 18, "Expelliarmus")
````

````python
harry.magic()
````
